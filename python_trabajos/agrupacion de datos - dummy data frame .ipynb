#AGREGACION DE DATOS POR CATEGORIA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gender = ["Mael", "Femail"]
income = ["Foor","Middle Class", "Rich"]
```

```python
n = 500
gender_data=[]
income_data = []
for i in range(0,500):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[:10]
```

```python
income_data[:10]
```

```python
height= 160 + 30 *np.random.randn(n)
weight = 65 + 15 * np.random.randn(n)
age = 30 + 12 + np.random.randn(n)
income= 18000 +3500 *np.random.randn(n)
```

```python
data=pd.DataFrame (
 
    {
        
        "Gender": gender_data,
        "Economic Status": income_data,
        "Height": height,
        "weight": weight,
        "Age": age,
        "Income": income
        
    }

)
```

```python
data.head(10)
```

```python
#agrupacion de datos
```

```python
grouped_gender= data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for name, groups in grouped_gender:
    print (name)
    print (groups)
```

```python
double_group=data.groupby(["Gender","Economic Status"])
```

```python
len(double_group)
```

```python
for name, group in double_group:
    print (name)
    print (group)
```

```python
##modificacion del grupo agragar valores que no interesen 
```

```python
double_group.aggregate(
 
     {
         "Income":np.sum,
         "Age":np.mean,
         "Height":np.std
         
     }

)
```

```python
## funcion lambda
```

```python
double_group.aggregate(

     {
         
         "Age" : np.mean,
         "Height": lambda h:(np.mean(h))/np.std(h)
     }

)
```

```python
double_group.aggregate([np.sum, np.mean ,np.std])
```

```python
## funcion lambda
```

```python
double_group.aggregate([lambda x :np.mean(x)/ np.std(x)])
```

```python
## filtrado de datos
```

```python
double_group["Age"].filter (lambda x : x.sum()>2400)
```

```python
## trasformacion de variables
```

```python
zscore= lambda x : (x - x.mean())/x.std()
```

```python
z_group=double_group.transform(zscore)
```

```python
plt.hist(z_group["Age"])
```

```python
## sirve para rellenar datos faltantes con el promedio
```

```python
fill_na_mean=lambda x : x.fillna ( x.mean())
```

```python
double_group.transform(fill_na_mean)
```
